```python
import matplotlib.pyplot as plt
```

x axis values

```python
x = [7,2,3]
# corresponding y axis values
y = [2,8,8]
```

plotting the points

```python
plt.plot(x, y)
```

naming the x axis

```python
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
```

giving a title to my graph

```python
plt.title('My second graph!')
```

function to show the plot

```python
plt.show()
```
